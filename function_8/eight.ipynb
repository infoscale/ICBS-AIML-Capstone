```python
print('X : ',  X)
print('Y : ', Y)
```

```python
import itertools as it
```

```python
dim = 8
X_grid = np.fromiter(it.chain(*it.product(np.linspace(0, 1, 9), repeat=dim)), dtype=float).reshape(-1,dim)
```

```python
def upper_confidance_bound(i_param, o_param):
    gpr = GaussianProcessRegressor()
    gpr.fit(i_param, o_param)
    mean, std = gpr.predict(X_grid, return_std = True)
    ucb = mean + 1.75 * std
    return ucb
```

```python
def get_next(upper_conf):
    idx_max = np.argmax(upper_conf)
    next_query = X_grid[idx_max]
    return next_query
```

```python
ucb = upper_confidance_bound(X,Y)
```

```python
print(get_next(upper_conf=ucb))
```
